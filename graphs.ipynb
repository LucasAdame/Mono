```python
import matplotlib.pyplot as plt
```

Dropout and ROC AUC Scores

```python
x = [0, 30,40, 50, 60, 70]
y1 = [0.92, 0.85,0.83, 0.82, 0.81, 0.73] #training
y2 = [0.65, 0.70,0.71, 0.71, 0.70, 0.67] #testing
plt.plot(x, y1, label= "banco de dados de treino")
plt.plot(x, y2, label = "banco de dados de teste")
plt.xlabel('Taxa de Dropout em percentagem')
plt.ylabel('AUC ROC Score')
plt.title('Efeito de diferentes taxas de dropout na rede neural')
plt.legend()
plt.show()
```

Number of epochs

```python
x = [1,2,3,5,10,15,30,50,100]
y1 = [0.76,0.79,0.79,0.79,0.81,0.82,0.82,0.82,0.82] #training
y2 = [0.69,0.69,0.71,0.73,0.73,0.71,0.71,0.71,0.71] #testing
plt.plot(x, y1, label= "banco de dados de treino")
plt.plot(x, y2, label = "banco de dados de teste")
plt.xlabel('NÃºmero de Epochs')
plt.ylabel('AUC ROC Score')
plt.title('Efeito de diferentes quantidades de epochs na rede neural')
plt.legend()
plt.show()
```
